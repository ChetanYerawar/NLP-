```python
all_Words = [nltk.word_tokenize(sent) for sent in train_data['combined']]
```

```python
word2vec = Word2Vec(all_Words, min_count=3)
```

```python
sim_words = word2vec.wv.most_similar('calm')
sim_words
```

```python
for i in train_data['answer_text'].head(100):
    option1 = ''
    option2 = ''
    option3 = ''
    for word in nltk.word_tokenize(i):
        try :
            sim_words = word2vec.wv.most_similar(word)
        except KeyError:
            continue
        option1 = option1 + ' ' + sim_words[0][0]
        option2 = option2 + ' ' + sim_words[1][0]
        option3 = option3 + ' ' + sim_words[2][0]
    
    print('**************************************************************************')
    print('correct answer : {}'.format(i))
    print('option 1 : {}'.format(option1))
    print('option 2 : {}'.format(option2))
    print('option 3 : {}'.format(option3))
    print('**************************************************************************')
```
